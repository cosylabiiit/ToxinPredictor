```python
threshold = 0.84
#code to remove highly correlated features
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    # corr_matrix = dataset.corr()  # Compute the correlation matrix
    # Drop unnamed column
    corr_matrix = pd.read_csv('correlation_matrix_MolToxPred.csv')
    # corr_matrix.to_csv('correlation_matrix_MolToxPred.csv',index=False)
    mask = ~corr_matrix.columns.str.match(r'^Unnamed|^$')
    corr_matrix = corr_matrix.loc[:, mask]
    print(corr_matrix)
    print("The shape of the correlation matrix is:", corr_matrix.shape)
    print("Total number of operations:", corr_matrix.shape[0] * corr_matrix.shape[1])
    
    index = 0
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:  # We are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # Getting the name of the column
                col_corr.add(colname)
            index += 1
    
    return col_corr
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report, roc_auc_score, precision_score, recall_score, roc_curve
from sklearn.model_selection import StratifiedKFold

# Load the dataset
df = pd.read_csv('merged_data_clean.csv')
categorical_columns = ['SMILES', 'Source']
df = df.drop(columns=categorical_columns)

Y = df['Toxicity']
X_pca_df = pd.read_csv('X_pca_clean.csv')

# Define the parameters for XGBoost
params = {
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'eta': 0.1,
    'max_depth': 6,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'random_state': 42
}

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

accuracy_list = []
f1_list = []
roc_auc_list = []
precision_list = []
recall_list = []

for train_index, test_index in skf.split(X_pca_df, Y):
    X_train, X_test = X_pca_df.iloc[train_index], X_pca_df.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    
    # Convert the data into DMatrix format required by XGBoost
    dtrain = xgb.DMatrix(X_train, label=Y_train)
    dtest = xgb.DMatrix(X_test, label=Y_test)

    # Train the model
    bst = xgb.train(params, dtrain, num_boost_round=100)

    # Make predictions
    Y_pred_prob = bst.predict(dtest)
    Y_pred = [1 if prob > 0.5 else 0 for prob in Y_pred_prob]

    # Calculate metrics
    accuracy = accuracy_score(Y_test, Y_pred)
    f1 = f1_score(Y_test, Y_pred)
    roc_auc = roc_auc_score(Y_test, Y_pred_prob)
    precision = precision_score(Y_test, Y_pred)
    recall = recall_score(Y_test, Y_pred)
    
    accuracy_list.append(accuracy)
    f1_list.append(f1)
    roc_auc_list.append(roc_auc)
    precision_list.append(precision)
    recall_list.append(recall)

# Print average metrics
print("Average Accuracy: ", np.mean(accuracy_list))
print("Average F1 Score: ", np.mean(f1_list))
print("Average ROC AUC Score: ", np.mean(roc_auc_list))
print("Average Precision: ", np.mean(precision_list))
print("Average Recall: ", np.mean(recall_list))

# Plot ROC curve for the last fold
fpr, tpr, _ = roc_curve(Y_test, Y_pred_prob)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_list[-1])
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report, roc_auc_score, precision_score, recall_score, roc_curve
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

accuracy_list = []
f1_list = []
roc_auc_list = []
precision_list = []
recall_list = []

for train_index, test_index in skf.split(X_pca_df, Y):
    X_train, X_test = X_pca_df.iloc[train_index], X_pca_df.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    
    # Standardize the data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    # Define the SVM model
    svm_model = SVC(kernel='rbf', probability=True, random_state=42)
    
    # Train the model
    svm_model.fit(X_train, Y_train)

    # Make predictions
    Y_pred_prob = svm_model.predict_proba(X_test)[:, 1]
    Y_pred = svm_model.predict(X_test)

    # Calculate metrics
    accuracy = accuracy_score(Y_test, Y_pred)
    f1 = f1_score(Y_test, Y_pred)
    roc_auc = roc_auc_score(Y_test, Y_pred_prob)
    precision = precision_score(Y_test, Y_pred)
    recall = recall_score(Y_test, Y_pred)
    
    accuracy_list.append(accuracy)
    f1_list.append(f1)
    roc_auc_list.append(roc_auc)
    precision_list.append(precision)
    recall_list.append(recall)

# Print average metrics
print("Average Accuracy: ", np.mean(accuracy_list))
print("Average F1 Score: ", np.mean(f1_list))
print("Average ROC AUC Score: ", np.mean(roc_auc_list))
print("Average Precision: ", np.mean(precision_list))
print("Average Recall: ", np.mean(recall_list))

# Plot ROC curve for the last fold
fpr, tpr, _ = roc_curve(Y_test, Y_pred_prob)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_list[-1])
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
